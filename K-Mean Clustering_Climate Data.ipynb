```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

#making dataframe

New_data_solar = pd.read_csv('irradiancedata.csv')
New_data_wind = pd.read_csv('wind.csv')
New_data_temp = pd.read_csv('tempdata.csv')

#clustering for irradiance data

kmeanssoalar = KMeans(n_clusters=5).fit(New_data_solar)
centroidssolar = kmeanssoalar.cluster_centers_
print(centroidssolar)

#clustering for wind data
kmeanswind = KMeans(n_clusters=5).fit(New_data_wind)
centroidswind = kmeanswind.cluster_centers_
print(centroidswind)

#clustring for tempreture data
kmeanstemp = KMeans(n_clusters=5).fit(New_data_temp)
centroidtemp = kmeanstemp.cluster_centers_
print(centroidtemp)

####plot
plt.scatter(New_data_solar['Rank'], New_data_solar['irradiance'], c= kmeanssoalar.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroidssolar[:, 0], centroidssolar[:, 1], c='red', s=50)
plt.show()

plt.scatter(New_data_wind['Rank'], New_data_wind['wind speed'], c= kmeanswind.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroidswind[:, 0], centroidswind[:, 1], c='red', s=50)
plt.show()


plt.scatter(New_data_temp['Rank'], New_data_temp['temp'], c= kmeanstemp.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroidtemp[:, 0], centroidtemp[:, 1], c='red', s=50)
plt.show()
```
